```python
#BUILDING A DECISION TREE CLASSFIER

#1 Load the Dataset
import pandas as pd
bank_df = pd.read_csv("bank_kaggle.csv")
bank_df.head()
```

```python
bank_df2 = pd.read_csv("bank_UCI.csv", sep =";")
bank_df2.head()
```

```python
#2 Work with X (Predictor Variables)
X_features = list(bank_df.columns)
X_features.remove('deposit')
X_features
```

```python
#3 Encoding the Categorial variables
encode_bank_df = pd.get_dummies(bank_df[X_features])
```

```python
encode_bank_df.head()
```

```python
#4 Set the X and Y
X =encode_bank_df
Y = bank_df.deposit.map(lambda x: int(x=='yes'))
```

```python
X
```

```python
Y
```

```python
#5 Split the Dataset
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.4, random_state = 20 )
```

```python
#6 Build the Decison Tree Classfier
from sklearn.tree import DecisionTreeClassifier

clf_tree = DecisionTreeClassifier(criterion = 'gini', max_depth = 4)
```

```python
clf_tree.fit(X_train, Y_train)
```

```python
#7 Accuracy Score
clf_tree.score(X_test, Y_test)
```

```python
# TASK 2
import pandas as pd
df = pd.read_csv("cereal-2.csv")
df.head()
```

```python

X_features = list(df.columns)
X_features.remove('fiber')
X_features
```

```python
encode_df = pd.get_dummies(df[X_features])
```

```python
encode_df.head()
```

```python
X = encode_df
Y = df.fiber.map(lambda x: int(x=='yes'))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.4, random_state = 20 )
```

```python
from sklearn.tree import DecisionTreeClassifier

tree = DecisionTreeClassifier(criterion = 'gini', max_depth = 4)
```

```python
tree.fit(X_train, Y_train)
```

```python
#7 Accuracy Score
tree.score(X_test, Y_test)
```

```python
from sklearn.tree import DecisionTreeRegressor  
regressor = DecisionTreeRegressor(random_state = 3)  
regressor.fit(X_train,Y_train)
```

```python
regressor.get_params()
```

```python
Y_pred = regressor.predict(X_test)
Y_pred
```
